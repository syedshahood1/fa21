```python
import random
from collections import deque
from time import perf_counter
import pandas as pd


class Node:

    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None


def generate_random_numbers(count, lower_bound, upper_bound):

    return random.sample(range(lower_bound, upper_bound + 1), count)


def build_tree(numbers):

    if not numbers:
        return None

    mid = len(numbers) // 2
    root = Node(numbers[mid])
    root.left = build_tree(numbers[:mid])
    root.right = build_tree(numbers[mid + 1:])
    return root


def bfs(root, goal):

    start_time = perf_counter()
    queue = deque([root])
    while queue:
        node = queue.popleft()
        if node.data == goal:
            return perf_counter() - start_time, node
        if node.left:
            queue.append(node.left)
        if node.right:
            queue.append(node.right)
    return perf_counter() - start_time, None


def dfs(root, goal):

    start_time = perf_counter()
    stack = [root]
    while stack:
        node = stack.pop()
        if node.data == goal:
            return perf_counter() - start_time, node
        if node.right:
            stack.append(node.right)
        if node.left:
            stack.append(node.left)
    return perf_counter() - start_time, None


def main():

    data_sets = [
        generate_random_numbers(1000, 1, 1000),
        generate_random_numbers(40000, 1, 40000),
        generate_random_numbers(80000, 1, 80000),
        generate_random_numbers(200000, 1, 200000),
        generate_random_numbers(1000000, 1, 1000000),
    ]

    results = []
    for i, data_set in enumerate(data_sets, 1):
        goal = data_set[-220]  # Access goal element efficiently using negative indexing
        sorted_data_set = sorted(data_set)
        tree = build_tree(sorted_data_set)
        bfs_time, bfs_node = bfs(tree, goal)
        dfs_time, dfs_node = dfs(tree, goal)

        results.append({
            "Dataset": f"Dataset {i}",
            "Number of Nodes": len(data_set),
            "Goal Node": goal,
            "BFS Time (s)": bfs_time,
            "BFS Found": bfs_node is not None,
            "DFS Time (s)": dfs_time,
            "DFS Found": dfs_node is not None,
        })

    # Create DataFrame
    df = pd.DataFrame(results)

    # Output DataFrame
    print(df)


if __name__ == "__main__":
    main()
```

```python
import heapq
from collections import deque


maze = [
    ['A', ' ', ' ', ' ', ' ', ' ', ' '],
    ['#', '#', '#', '#', ' ', '#', ' '],
    [' ', ' ', ' ', '#', ' ', ' ', '#'],
    ['#', '#', ' ', '#', '#', ' ', ' '],
    ['#', ' ', ' ', ' ', '#', '#', ' '],
    [' ', ' ', '#', ' ', ' ', ' ', 'Y'],
]

# Define start and goal nodes
start_node = (0, 0)  # A
goal_node = (5, 6)   # Y


movements = [(-1, 0), (1, 0), (0, -1), (0, 1)]


def heuristic(node):

    return abs(node[0] - goal_node[0]) + abs(node[1] - goal_node[1])


def a_star_search():

    open_list = []
    heapq.heappush(open_list, (heuristic(start_node), 0, start_node, []))
    closed_set = set()

    while open_list:
        _, cost, current_node, path = heapq.heappop(open_list)

        if current_node == goal_node:
            return path + [current_node]

        if current_node in closed_set:
            continue

        closed_set.add(current_node)

        for move in movements:
            next_node = (current_node[0] + move[0], current_node[1] + move[1])

            if 0 <= next_node[0] < len(maze) and 0 <= next_node[1] < len(maze[0]) \
                    and maze[next_node[0]][next_node[1]] != '#':
                next_cost = cost + 1
                heapq.heappush(open_list, (next_cost + heuristic(next_node), next_cost, next_node, path + [current_node]))

    return None


def print_path(path):

    for node in path:
        row, col = node
        maze[row][col] = 'X'  # Mark the path with X
    for row in maze:
        print(' '.join(row))


def main():

    path = a_star_search()
    if path:
        print("Path found:")
        print_path(path)
    else:
        print("No path found.")


if __name__ == "__main__":
    main()
```

```python
import math


MAX_DEPTH = 3


class Node:
    def __init__(self, value=None):
        self.value = value
        self.children = []

    def add_child(self, child):
        self.children.append(child)


def minimax(node, depth, alpha, beta, maximizing_player):
    if depth == 0 or not node.children:
        return node.value

    if maximizing_player:
        value = -math.inf
        for child in node.children:
            value = max(value, minimax(child, depth - 1, alpha, beta, False))
            alpha = max(alpha, value)
            if alpha >= beta:
                break
        return value
    else:
        value = math.inf
        for child in node.children:
            value = min(value, minimax(child, depth - 1, alpha, beta, True))
            beta = min(beta, value)
            if alpha >= beta:
                break
        return value


def alpha_beta_pruning(root):
    return minimax(root, MAX_DEPTH, -math.inf, math.inf, True)


def main():

    root = Node()
    root.add_child(Node(3))
    root.add_child(Node(5))
    root.add_child(Node(6))
    root.children[0].add_child(Node(9))
    root.children[0].add_child(Node(1))
    root.children[1].add_child(Node(2))
    root.children[1].add_child(Node(0))
    root.children[2].add_child(Node(7))
    root.children[2].add_child(Node(8))

    best_move = alpha_beta_pruning(root)
    print("Best move:", best_move)


if __name__ == "__main__":
    main()
```
